# 1. 베이지안 추론

'밑바닥부터 시작하는 데이터과학' (한빛미디어)를 정리했음을 밝힙니다.

### 문제 정의와 베타분포
동전이 하나 있습니다. 우리는 이 동전이 공평한 동전인지 아닌지 알아보고 싶습다면, 이동전을 던졌을 때 앞면이 나올 확률이 p라고 하면 이 p가 어떤 값인지 추정해 보면됩니다.  

동전던지기 실험은 (이항분포)를 따릅니다. 이항분포란 성공확률이 p이고 그 결과가 성공 혹은 실패 뿐인 실험을 n번 반복시행할 때 성공횟수의 분포를 가리킵니다.  

이항분포 파라메터 p의 사전확률은 베타분포를 따릅니다. 여기서 사전확률이란 데이터를 관측하기 전 가설로 세운 확률을 뜻합니다.  

이 때문에 우리가 알고 싶은 앞면이 나올 확률 p는 베타분포를 사전분포로 사용하게 됩니다. 베타분포도 정규분포처럼 중심이 높은 포물선 형태를 갖습니다. 베타분포의 파라메터는 a와 b이며 그 중심은 다음과 같습니다.  

- a/(a+b)

p의 사전확률과 관련해 a는 성공(앞면) b는 실패(뒷면)과 연관이 있습니다, a,b의 크기는 믿음의 크기와 관련이 있습니다. 만약 동전에 대해 어떤 선입견도 취하고 싶지 않다면 a,B를 모두 1로 정하면 됩니다. 앞면이 55%인경우로 나타난다고 굳게 믿고 싶으면 a를 55로 B를 45로 가정할 수도 있습니다.  

이항분포 파라메터 p의 사후 확률 역시 베타분포를 따릅니다. 여기에서 사후확률이란 데이터를 확인한 이후의 가설 확률을 가리킵니다. 바로 우리가 알고 싶은 값입니다. 사후확률은 사전 확률의 업데이트 버전 정도로 이해하면 됩니다.  

### 추론 과정

동전을 여러번 던져서 앞면이 h번 뒷면이 t번 나왔다고 칩시다, 베이즈 규칙을 활용해 유도하면 p의 사후확률 분포는 파라메터가 a+h,B+t인 베타분포를 따른다고 합니다.  

이제 동전을 10번 던져서 앞면을 3번 관측했다고 가정해 보겠습니다. 실험 대상 동전이 공평한 동전이라고 생각해 a,B를 모두 1로 두었다고 하면 앞면이 나올 확률p의 사후 분포는 파라메터가 1+3, 1+7인 베타분포가 될것입니다.  

공평한 동전이라는 생각이 다소 강하게 들어 a,B를 각각 20으로 정했다면 p의 사후분포는 Beta(20+3,20+7)이 됩니다. 즉,  동전이 뒷면으로 살짝 편향되었다는 것으로 믿음이 바뀔 것을 의미합니다.  

앞면이 잘 나오는 동전이라는 확신 떄문에 p의 사전분포를 Beta(30,10)로 두었다면 사후 분포는 Beta(30+3,10+7)이 됩니다. 이 경우 동전이 아직도 앞면이 편햔되어 있다고 믿지만, 처음 생각했던 것 보다는 덜 강하게 믿는걸 뜻합니다.  

데이터 관측 후 a,B 변화에 따른 p의 사후분포 변화는 다음 그림과 같습니다. 예컨데 Beta(4,8)의 경우 중심이 0.33입니다. 이는 p의 사후확률이 0.33일 가능성이 제일 높다는 뜻입니다.   

![](http://i.imgur.com/P7L4A8W.png)

동전을 많이 던져볼 수록 관측 전 가정한 사전분포는 점점 의미가 없어집니다.  예를 들어 편향된 동전이라는 사전 믿음이 아무리 강했을지라도 동전을 2천번 던져서 앞면에 천번이 나왔다면 생각을 바꿀 수 밖에 없을것입니다. 데이터가 충분하다면 서로 다른 사전 확률을 가지고 시작한다고 해도 동일한 사후확률로 수렴하는 경향이 있습니다.  

이러한 과정을 통해 p가 가지는 값을 확률적으로 추정할 수가 있게 됩니다. 이를 베이지안 추론이라고 합니다. 이는 p-value를 이용한 가설검정 결과와는 접근 방식이 근본적으로 다릅니다. 예컨데 다음과 같습니다.  

- 베이지안 추론 : 관측한 데이터와 사전분포를 고려해볼 때 동전의 앞면이 나올 확률이 49~51%일 경우는 5%밖에 되지 않는다.

- 가설검정 : 동전이 공평하다면 이렇게 편향된 데이터를 관측할 경우는 5%밖에 되지 안흔다.

베이지안 추론 과정은 사전에 가설을 세운뒤 실험(데이터 관측)결과로 가설을 조금씩 업데이트 하면서 완성이 됩니다. 평소 우리가 경험적으로 체감하고 있는 확류들(예컨데 전철에서 자리가 날 확률, 휴강을 할 확률 등등)은 사실 베이지안 추론과 본질상 다르지 않습니다.

### 시각화
위 그림을 만드는 파이썬 코드입니다.

```python
import math
from matplotlib import pyplot as plt
def B(alpha, beta):
    return math.gamma(alpha) * math.gamma(beta) / math.gamma(alpha + beta)
def beta_pdf(x, alpha, beta):
    # [0, 1] 구간 밖에서는 밀도가 없음
    if x < 0 or x > 1:
        return 0
    return x ** (alpha - 1) * (1 - x) ** (beta - 1) / B(alpha, beta)

xs = [x / 100.0 for x in range(0,100)]
plt.plot(xs,[beta_pdf(x,alpha=4,beta=8) for x in xs],'-',label='Beta(4,8)')
plt.plot(xs,[beta_pdf(x,alpha=23,beta=27) for x in xs],'--',label='Beta(23,27)')
plt.plot(xs,[beta_pdf(x,alpha=33,beta=17) for x in xs],':',label='Beta(33,17)')
plt.legend()
plt.title('Beta pdfs')
plt.show()
```
